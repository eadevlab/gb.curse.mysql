## Практическая работа №5 Теория вероятностей и математическая статистика

#### Задание №1
Известно, что генеральная совокупность распределена нормально
со средним квадратическим отклонением, равным 16.
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,
если выборочная средняя M = 80, а объем выборки n = 256.

т.к. указан тип распределения генеральной совокупности и известно среднее квадратичное отклонение = 16,можно воспользоваться формулой
$T_{1,2} = \bar X \pm \frac{t_{\gamma} \cdot \sigma}{\sqrt{n}}$, где<br>
$\bar X = M$ - среднее значение<br>
$n$ - объём выборки<br>
$\sigma$ - средне квадратичное отклонение<br>
$t_{\gamma}$ - коэфициент доверия $2\Phi(t_{\gamma}) = \gamma => \Phi(t_{\gamma}) = \frac{\gamma}{2}$<br>
$\Phi(t_{\gamma})$ - функция Лапласа<br>
$\Phi(t_{\gamma}) = \frac{0.95}{2} => \Phi(t_{\gamma}) = 0.475$, заглянув в таблицу Лапласа выясним что аргумент функции $t_{\gamma} = 1.96$


$$T_{1,2} = 80 \pm \frac{1.96 \cdot 16}{\sqrt{256}} = 80 \pm 1.96$$
$$T_1 = 80 - 1.96 = 78,04$$
$$T_2 = 80 + 1.96 = 81.96$$

**Ответ:**<br>
доверительный интервал для оценки математического ожидания a с надежностью 0.95 равен **(78.04; 81.96)**

#### Задание №2
В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью,
получены опытные данные:
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей,
оценить истинное значение величины X при помощи доверительного интервала, покрывающего это
значение с доверительной вероятностью 0,95.

если генеральное стандартное отнклонение неизвестно, то интервал можно построить по формуле
$T_{1,2} = \bar X \pm \frac{t_{\gamma} \cdot \sigma}{\sqrt{n}}$, где
$\bar X$ - среднее по выборке<br>
$t_{\gamma}$ - коэфициент доверия расчитанный с помощью распределения Стьюдента<br>
$\sigma$ - несмещенная оценка генерального стандартного отклонения

Для начала найдём выборочные характеристики, среднее и диссперсию
$$\bar X = \frac{\sum_{i=1}^{n} x_i}{n} = \frac{6.9 + 6.1 + 6.2 + 6.8 + 7.5 + 6.3 + 6.4 + 6.9 + 6.7 + 6.1}{10} = 6.59$$

$$D(X) = \frac{\sum_{i=1}^{n} (x_i - \bar X)^2}{n}$$
$$\sum_{i=1}^{n} (x_i - \bar X)^2 = (6.9-6.59)^2 + (6.1-6.59)^2 + (6.2-6.59)^2 + (6.8-6.59)^2 + (7.5-6.59)^2 + (6.3-6.59)^2 + (6.4-6.59)^2 + (6.9-6.59)^2 + (6.7-6.59)^2 + (6.1-6.59)^2 = 1.829$$
$$D(X) = \frac{1.829}{10} = 0.1829$$

$$\sigma = \sqrt{\frac{n}{n-1}\cdot D(X)} = \sqrt{\frac{10}{9}\cdot 0.1829} = 0.4508$$

согласно таблице распределения Стьюдента для доверительной вероятности $\gamma = 0.95$ и объёму выборки $n = 10$ соответствует коэффициент доверия:
$t_{\gamma} = 2.26$


Теперь можно найти доверительный интервал
$$T_{1,2} = \bar X \pm \frac{t_{\gamma}\cdot \sigma}{\sqrt{n}} = 6.59 \pm \frac{2.26 \cdot 0.4508 }{\sqrt{10}} = 6.59 \pm 0.3222$$
$$T_{1} = 6.59 - 0.3222 = 6.2678$$
$$T_{1} = 6.59 + 0.3222 = 6.9122$$

**Ответ: (6.2678; 6.9122)**

#### Задание №3
Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.

Сформулируем гипотезы:<br>
$H_0$ - нулевая гипотеза $\mu = \mu_{H_0} = 17$
$H_1$ - альтернативная гипотеза $\mu > \mu_{H_0}$
где, $\mu - среднее выборки$,а $\mu_{H_0} - среднее генеральной выборки$
т.к. известно среднее значение и дисперсия можно воспользоваться z-тестом (z-критерием Фишера)

$$z_{\bar X} = \frac{\bar X - \mu_{H_0}}{\sigma/\sqrt{n}} = \frac{17.5 - 17}{\sqrt{4}/\sqrt{100}} = \frac{0.5}{0.2} = 2.5$$

Для $\alpha = 0.05$ критическое значение Z-критерия равно 1.96. т.к. полученное $z_{\bar X} > z_{крит}$, нулевая гипотеза отвергается

**Ответ: нулевая гипотеза $H_0: \mu_{H_0} = 17$ отвергается, при $\alpha = 0.05$**

#### Задание №4
Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

Сформулируем гипотезы:<br>
$H_0$ - нулевая гипотеза $\mu = \mu_{H_0} = 200$<br>
$H_1$ - альтернативная гипотеза $\mu != \mu_{H_0}$<br>
т.к. неизвестны средние значения и дисперсия можно использовать t-критерий Стьюдента, но при этом использовать двустороний критерий значимости $\alpha = 1 - 0.99 = 0.01$

$$t = \frac{\bar X -\mu_{H_0}}{\sigma_X/\sqrt{n}}$$

где,
$\bar X$ - Среднее значение по выборке
$\sigma_X$ - Несмещенная оценка дисперсии

$$\bar X = \frac{\sum_{i=1}^{n} x_i}{n} = \frac{202 + 203 + 199 + 197 + 195 + 201 + 200 + 204 + 194 + 190}{10} = 198.5$$

$$\sigma_X^2 = \frac{\sum_{i=1}^{n} (x_i - \bar X)^2}{n-1} = \frac{(202-198.5)^2 + (203-198.5)^2 + (199-198.5)^2 + (197-198.5)^2 + (195-198.5)^2 + (201-198.5)^2 + (200-198.5)^2 + (204-198.5)^2 + (194-198.5)^2 + (190-198.5)^2}{9}$$

$$\sigma_X = \sqrt{19.8333} = 4.4535$$

$$t = \frac{\bar X -\mu_{H_0}}{\sigma_X/\sqrt{n}} = \frac{198.5 - 200}{4.4535 / \sqrt{10}} = \frac{−1.5}{1,4083} = −1.0651$$

Осталось найти табличное значение t-критерия, для $\alpha/2 = 0.005$<br>
$t_{таб} = 3.2498$<br>
$|t| < t_{таб}$, значит нулевая гипотеза принимается на уровне значимости $\alpha = 0.01$

**Ответ: Нулевая гипотеза $\mu = 200 $ принимается на уровне значимости $\alpha = 0.01$**
