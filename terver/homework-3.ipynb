#### Задание №1
##### Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

Среднее арифметическое:
$$\bar x = \frac{100+80+75+77+89+33+45+25+65+17+30+24+57+55+70+75+65+84+90+150}{20} = \frac{1306}{20} = 65.3$$

Среднее квадратичное отклонение:
$$\sigma=\sqrt{\frac{\displaystyle\sum^n_{i=1}(x_{i}-\bar x)^2}{n}} = \sqrt{\frac{(100-65.3)^2+(80-65.3)^2+(75-65.3)^2+(77-65.3)^2+(89-65.3)^2+(33-65.3)^2+(45-65.3)^2+(25-65.3)^2+(65-65.3)^2+(17-65.3)^2+(30-65.3)^2+(24-65.3)^2+(57-65.3)^2+(55-65.3)^2+(70-65.3)^2+(75-65.3)^2+(65-65.3)^2+(84-65.3)^2+(90-65.3)^2+(150-65.3)^2}{20}}$$
$\sigma = \sqrt{\frac{19002,2}{20}} = \sqrt{950,11} = 30,8239$

Смещенная оценка дисперсии
$$\sigma^2 = \frac{\displaystyle\sum^n_{i=1}(x_{i}-\bar x)^2}{n} = 30.8239^2 = 950.11$$
Несмещенная оценка дисперсии
$$\sigma^2 = \frac{\displaystyle\sum^n_{i=1}(x_{i}-\bar x)^2}{n-1} = \frac{19002.2}{19} = 1000.1158$$

**Ответ:**
- Среднее арифметическое: **65.3**
- Среднее квадратичное отклонение: **30.8239**
- Смещенная оценка дисперсии: **950.11**
- Несмещенная оценка дисперсии: **1000.1158**

#### Задание №2
##### В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Для данного события (А) есть 3 варианта развития событий
1) $A_1$ - из первой корзины вытащили 2 белых мяча, из второй - 1
2) $A_2$ - из первой корзины вытащили 1 белый мяч, из второй - 2
3) $A_3$ - из первой корзины вытащили 0 белых мячей, из второй - 3
$P(A) = P(A_1 + A_2 + A_3) = P(A_1) + P(A_2) + P(A_3)$

Для удобства посчитаем число элементарных исходов
**Для первой корзины**
$C^{2}_8 = \frac{8!}{2!(8-2)!} = \frac{8!}{2!6!} = \frac{7*8}{1*2} = 7*4 = 28$
**Для второй корзины**
$C^4_{12} = \frac{12!}{4!(12-4)!} = \frac{12!}{4!8!} = \frac{9*10*11*12}{1*2*3*4} = 3*5*11*3 = 495$

$$P(A_1) = \frac{C^2_5}{C^2_8}\cdot\frac{C^1_5\cdot C^3_7}{C^4_{12}} = \frac{\frac{5!}{2!3!}}{28}\cdot \frac{\frac{5!}{4!}\cdot \frac{7!}{3!4!}}{495} = \frac{\frac{4*5}{2}}{28} \cdot \frac{5\cdot \frac{5*6*7}{2*3}}{495} = 0.3571*0.3535 = 0.1262$$

$$P(A_2) = \frac{C^1_5\cdot C^1_3}{C^2_8}\cdot\frac{C^2_5\cdot C^2_7}{C^4_{12}} = \frac{\frac{5!}{1!4!}\cdot \frac{3!}{1!2!}}{28} \cdot \frac{\frac{5!}{2!3!}\cdot \frac{7!}{2!5!}}{495} = \frac{5*3}{28} \cdot \frac{\frac{4*5*6*7}{2*2}}{495} = \frac{15}{28} \cdot \frac{210}{495} = 0.5357*0.4242 = 0.2272$$

$$P(A_3) = \frac{C^0_5\cdot C^2_3}{C^2_8} \cdot \frac{C^3_5\cdot C^1_7}{C^4_{12}} = \frac{\frac{3!}{2!1!}}{28} \cdot \frac{\frac{5!}{3!2!}\cdot \frac{7!}{1!6!}}{495} = \frac{3}{28} \cdot \frac{10*7}{495}$ = 0.1071 * 0.1414 = 0.01514$$

$$P(A) = 0.1262 + 0.2272 + 0.0151 = 0.3685 = 36.85\%$$

**Ответ: 0.3685 (36.85\%)**

#### Задание №3
##### В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию.
Какова вероятность, что он учится:
a). на факультете A
б). на факультете B
в). на факультете C?

Исходя из условия задачи расчитаем вероятность того что студент учится на одном из факультетов,
Полная вероятность того что студент учится на одном из факультетов равна
$P(A)+P(B)+P(C) = 1$, т.к. P(A) = P(B) и P(C) = P(A+B) = 2P(A) можно записать по другому
$P(A) + P(A) + 2P(A) = 1$
$P(A) = \frac{1}{4}, P(B) = \frac{1}{4}, P(C) = \frac{1}{2}$


Для решения задачи воспользуемся формулой Байеса.
Для начала расчитаем полную вероятность
$$P(E) = P(A)*P(E|A) + P(B)*P(E|B) + P(C)*P(E|C)$$,
где
- P(E) - сессию сдали
- P(E|(A|B|C)) - вероятность сдачи сесси студентом факультета A,B,C
- P(A), P(B), P(C) - вероятность того что студент учится на факультете A,B,C
$$P(E) = \frac{1}{4}*0.8 + \frac{1}{4}*0.7 + \frac{1}{2}*0.9 = 0,825$$

**Вероятность того что студент сдавший сессию учится на факультете А**
$$P(A|E) = \frac{P(A)\cdot P(E|A)}{P(E)} = \frac{0.25*0.8}{0.825} = 0.2424 (24.24\%)$$
**Вероятность того что студент сдавший сессию учится на факультете B**
$$P(B|E) = \frac{P(B)\cdot P(E|B)}{P(E)} = \frac{0.25*0.7}{0.825} = 0.2121 (21.21\%)$$
**Вероятность того что студент сдавший сессию учится на факультете C**
$$P(C|E) = \frac{P(C)\cdot P(E|C)}{P(E)} = \frac{0.5*0.9}{0.825} = 0.5455 (54.55\%)$$

**Ответ:**
**a). 0.2424 (24.24\%)**
**б). 0.2121 (21.21\%)**
**в). 0.5455 (54.55\%)**

#### Задание №4
##### Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25.
Какова вероятность того, что в первый месяц выйдут из строя:
а). все детали
б). только две детали
в). хотя бы одна деталь
г). от одной до двух деталей?

a) В первый месяц работы из строя вышли все детали
$P = P(A_1)*P(A_2)*P(A_3) = 0.1*0.2*0.25 = 0.005 = 0.5\%$

---

б) В первый месяц работы из строя вышли только две детали
Данное событие может случится только в 3х случаях:
- Деталь 1 и 2 вышли из строя, а 3 не вышла
- Деталь 2 и 3 вышли из строя, а 1 не вышла
- Деталь 1 и 3 вышли из строя, а 2 не вышла
тогда вероятность можно посчитать как
$$P = P(A_1)*P(A_2)*P(\bar A_3) + P(\bar A_1)*P(A_2)*P(A_3) + P(A_1)*P(\bar A_2)*P(A_3)$$
$$= 0.1*0.2*(1-0.25)+(1-0.1)*0.2*0.25+0.1*(1-0.2)*0.25 = 0.08 = 8\%$$

---

в) В первый месяц из строя вышла хотябы одна деталь
пойдём от обратного и найдём вероятность того что из строя не выйдет ни одной детали, после чего отнимем полученное значение от 1
$P = 1 - \bar P = 1 - P(\bar A_1)*P(\bar A_2)*P(\bar A_3) = 1 - (1-0.1)*(1-0.2)*(1-0.25) = 0.46 = 46\%$

---

г) Из строя выйдет от одной до двух деталей
данное событие наступит когда вышла из строя одна деталь или из строя вышла две детали, т.е. можно рассматривать это как сумму вероятностей двух событий. $P = P_1 + P_2$, вероятность того что из строя выйдет 2 детали мы уже знаем, найдём чему равна вероятность выхода из строя одной детали.
$$P_1 = P(\bar A_1)*P(\bar A_2)*P(A_3) + P(\bar A_1)*P(A_2)*P(\bar A_3) + P(A_1)*P(\bar A_2)*P(\bar A_3) = $$
$$ = (1-0.1)*(1-0.2)*0.25 + (1-0.1)*0.2*(1-0.25) + 0.1*(1-0.2)*(1-0.25) = 0.375$$
$$P = P_1 + P_2 = 0.08 + 0.375 = 0.455 = 45.5\%$$

---

#### Ответ:
а) Вероятность того что из строя выйдут все детали равна **0.005 (0.5\%)**
б) Вероятность того что из строя выйдут 2 детали равна **0.08 (8\%)**
в) Вероятность того что из строя выйдут хотя бы одна деталь равна **0.46 (46\%)**
г) Вероятность того что из строя выйдут от одной до двух деталей равна **0.455 (45.5\%)**
