## Практическая работа №2 Теория вероятностей и математическая статистика

#### Задание №1
##### Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

Среднее арифметическое:
$$\bar x = \frac{100+80+75+77+89+33+45+25+65+17+30+24+57+55+70+75+65+84+90+150}{20} = \frac{1306}{20} = 65.3$$

Среднее квадратичное отклонение:
$$\sigma=\sqrt{\frac{\displaystyle\sum^n_{i=1}(x_{i}-\bar x)^2}{n}} = \sqrt{\frac{(100-65.3)^2+(80-65.3)^2+(75-65.3)^2+(77-65.3)^2+(89-65.3)^2+(33-65.3)^2+(45-65.3)^2+(25-65.3)^2+(65-65.3)^2+(17-65.3)^2+(30-65.3)^2+(24-65.3)^2+(57-65.3)^2+(55-65.3)^2+(70-65.3)^2+(75-65.3)^2+(65-65.3)^2+(84-65.3)^2+(90-65.3)^2+(150-65.3)^2}{20}}$$
$\sigma = \sqrt{\frac{19002,2}{20}} = \sqrt{950,11} = 30,8239$

Смещенная оценка дисперсии
$$\sigma^2 = \frac{\displaystyle\sum^n_{i=1}(x_{i}-\bar x)^2}{n} = 30.8239^2 = 950.11$$
Несмещенная оценка дисперсии
$$\sigma^2 = \frac{\displaystyle\sum^n_{i=1}(x_{i}-\bar x)^2}{n-1} = \frac{19002.2}{19} = 1000.1158$$

_Ответ:_
- Среднее арифметическое: _65.3_
- Среднее квадратичное отклонение: _30.8239_
- Смещенная оценка дисперсии: _950.11_
- Несмещенная оценка дисперсии: _1000.1158_
