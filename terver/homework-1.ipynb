## Практическая работа №1 Теория вероятностей и математическая статистика

#### Задание №1
Из колоды в 52 карты извлекаются случайным образом 4 карты.
a) Найти вероятность того, что все карты – крести.
б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

а)
т.к. в колоде 13 крест вероятность успешного исхода будет $C^4_{13}$, всего исходов $C^4_{52}$
$$P = \frac{C^4_{13}}{C^4_{52}} = \frac{\frac{13!}{4!(13-4)!}}{\frac{52!}{4!(52-4)!}} = \frac{\frac{13!}{4!9!}}{\frac{52!}{4!48!}} = \frac{13!4!48!}{4!9!52!} = \frac{13!48!}{9!52!} = \frac{10*11*12*13}{49*50*51*52} = \frac{17160}{6497400} = 0.0026 = 0.26\%$$

##### Ответ: 0.0026

б)
Решение задач с "хотя бы" делается от противного. т.е. нужно найти вероятность того что среди карт не будет ниодного туза и вычесть полученную вероятность от 1
$$
P = 1 - \frac{C^4_{48}}{C^4_{52}} = 1 - \frac{48!48!}{44!52!} = 1 - \frac{45*46*47*48}{49*50*51*52} = 1 - \frac{4669920}{6497400} = 1 - 0.718736725 = 0,2812 = 28.12\%
$$
##### Ответ: 0.2812

#### Задание №2
На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

$$P = \frac{1}{C^3_{10}} = \frac{1}{\frac{10!}{3!(10-3)!}} = \frac{3!7!}{10!} = \frac{1*2*3}{8*9*10} = \frac{6}{720} = 0,008(3) = 0.83\%$$
##### Ответ: 0.083

#### Задание №3

В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

$$P = \frac{C^3_{9}}{C^3_{15}} = \frac{\frac{9!}{3!(9-3)!}}{\frac{15!}{3!(15-3)!}} = \frac{9!12!}{6!15!} = \frac{7*8*9}{13*14*15} = \frac{504}{2730} = 0.1846 = 18.46\%$$
Ну или можно по другому, вероятность вытащить одну окрашенную деталь из 15 равна 9/15, вторую из 14 - 8/14 и 3ю - 7/13, теперь остаётя только перемножить эти вероятности
$$P = \frac{9}{15}\cdot\frac{8}{14}\cdot\frac{7}{13} = \frac{504}{2730} = 0.1846 = 18.46\%$$
##### Ответ: 0.1846

#### Задание №4
В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

$$P = \frac{1}{C^2_{100}} = \frac{2!(100-2)!}{100!} = \frac{1*2}{99*100} = \frac{2}{9900} = 0.0002 = 0.02\%$$
##### Ответ: 0.0002

#### Задание №5
На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен:
a). первым спортсменом
б). вторым спортсменом
в). третьим спортсменом.

Для начала необходимо найти полную вероятность того что в мишень попали
$$P(A) = P(B_1)P(A|B_1) + P(B_2)P(A|B_2) + P(B_3)P(A|B_3)$$, где
$P(A)$ - в мишень попали
$P(B_n)$ - выстрел произвёл n-й спортсмен ($\frac{1}{3}$)
$P(A|B_n)$ - n-й спортсмен попал в мишень

$$P(A) = \frac{1}{3}*0.9 + \frac{1}{3}*0.8 + \frac{1}{3}*0.6 = 0,7(6)$$
вот теперт можно найти отдельные вероятности
$$P(B_n|A) = \frac{P(B_n)P(A|B_n)}{P(A)}$$

$$P(B_1|A) = \frac{1/3*0.9}{0.7666} = 0.3913$$
$$P(B_2|A) = \frac{1/3*0.8}{0.7666} = 0,3478$$
$$P(B_3|A) = \frac{1/3*0.6}{0.7666} = 0,2608$$

##### Ответ: а) 0.3913, б) 0,3478, в) 0,2608
